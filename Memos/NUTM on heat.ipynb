# General UTM solution on the finite interval

Say we want to solve 
\begin{align*}
q_t+\omega(-i\partial_x)q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= q_0, &x\in[0,L],\\
\partial_x^j q(0,t) &= g_j(t), & j=0,\dots,n-1,\\
\partial_x^j q(L,t) &= h_j(t), & j=0,\dots,n-1,
\end{align*}
where $\omega(k)\in\mathbb{P}^{n}$. Note that the boundary conditions are not neccesarily imposed, but some should follow from others through a procedure that will be given by example.
*** Work out that the solution is given by ***

\begin{align*}
q(x,t)  = \frac{1}{2π}\int_{-\infty}^{\infty}e^{ikx-\omega(k)t}\hat{q}_0(k)\mathrm{d}k &- \frac{1}{2π}\int_{∂ D^+}e^{ikx-\omega(k)}\tilde{g}(k,t)\mathrm{d}k\\
& - \frac{1}{2π}\int_{∂ D^-}e^{-ik(L-x)-\omega(k)}\tilde{h}(k,t)\mathrm{d}k,
\end{align*}
where $\hat{q}_0=\int_{0}^{L}e^{-ikx}q_0\mathrm{d}x$, $∂ D^+\{k|\mathcal{I}m(k) ≥ 0 \& \omega(k)=0\}$ and $∂ D^-\{k|\mathcal{I}m(k) ≤ 0 \& \omega(k)=0\}$. We now also have $\tilde{g} = \sum_{j=0}^{n-1}c_j(k)\tilde{g}_j(\omega(k),t)$, with $\tilde{g}_j=\int_{0}^t e^{ks}g(s)\mathrm{d}s$ and respectively the same for $\tilde{h}$, we also define the coeficients through
$$ \sum_{j=0}^{n-1}c_j(k)\partial_x^jq(x,t) = i\left(\frac{\omega(k)-\omega(l)}{k-l}\right)\Big|_{l=-i\partial_x}q(x,t).$$

# Heat equation with homogeneous boundary condition.
To get started, let's deal with the heat equation with Homogeneous boundary conditions, here we have $\omega(k)=k^2$, $g_0≡0$ and $h_0≡0$:
\begin{align*}
q_t-\partial_{xx}q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= q_0, &x\in[0,L],\\
q(0,t) &= 0, & t ∈ [0,T],\\ 
q(L,t) &= 0, & t ∈ [0,T]
\end{align*}
We first need to find $\tilde{g}$ and $\tilde{h}$, but for this we need to find $g_1$ and $h_1$ and the coefficients $c_j(k)$. The latter is easy, since
$$c_0(k)q(x,t)+c_1(k)\partial_x q(x,t)=i\left(\frac{k^2-l^2}{k-l}\right)\Big|_{l=-i\partial_x}q(x,t)=ikq(x,t)+\partial_xq(x,t)~⇒~c_0 = ik~\&~c_1= 1.$$
The more difficult part is to find $g_1$ and $h_1$. For this purpose we use the global relation *** it's easy to get, but I'm gonna ignore that for now***
$$\hat{q}_0(k)-\tilde{g}(\omega(k),t)+e^{-ikL}\tilde{h}(\omega(k),t)=e^{\omega(k)t}\hat{q}(k,t),~~k\in\mathbb{C}.$$
Using invariances under $\omega$, the previous coefficients and $h_0≡ g_0≡ 0$, we get
\begin{align*}
\hat{q}_0(k)-\tilde{g}_1(k^2,t)+e^{ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(k,t)\\
\hat{q}_0(-k)-\tilde{g}_1(k^2,t)+e^{-ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(-k,t)\\
\end{align*}
Solving the system to get $\tilde{g}_1$ and $\tilde{h}_1$ gives us for $\Delta(k) = e^{ikL}-e^{-ikL}$,
\begin{align*}
\tilde{g}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right)-\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)e^{-ikL}-e^{k^2t}\hat{q}(k)e^{ikL}\right),\\
\tilde{h}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right)-\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)-e^{k^2t}\hat{q}(k)\right).
\end{align*}
*** Show that the latter of the two will eventually be equal to zero when we use the global relation and use the inverse fourier transform on it. This is true because of analyticity and the Jordan's lemma to prove that the whole upper (or lower) half plane is nicely behaving such that that arc is zero as it's radius goes to infinity.***
Therefore,
\begin{align*}
&\tilde{g}(k,t)=\tilde{g}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right),\\
&\tilde{h}(k,t)=\tilde{h}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right).
\end{align*}
The solution is then of the form 
\begin{align*}
q(x,t)  = \frac{1}{2π}\int_{-\infty}^{\infty}e^{ikx-\omega(k)t}\hat{q}_0(k)\mathrm{d}k &- \frac{1}{2π}\int_{∂ D^+}e^{ikx-\omega(k)t}\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right)\mathrm{d}k\\
& - \frac{1}{2π}\int_{∂ D^-}e^{-ik(L-x)-\omega(k)t}\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right)\mathrm{d}k.
\end{align*}
Numerical approximations go as follows, note that we put an emphasis on the asymptotics for $t→0$ as we aim to build a timestepping method.

# First integral
$$
I = \int_{-\infty}^{\infty}e^{ik x - \omega(k)t}\hat{q}_0(k)\mathrm{d}k.
$$
Note that if we have
$$
\tilde{q}(x) = q(φ^{-1}(x)), \text{ with }\tilde{q}:[-1,1]\to\mathbb{R}\text{ and } \varphi^{-1}:[-1,1]\to[0,L]: x ↦ \frac{L}{2}(x+1),
$$
then
$$
\mathcal{F}(q)[k] = \frac{L}{2}e^{-ik\frac{L}{2}} \mathcal{F}(\tilde{q})\left[\frac{L}{2}k\right].
$$
Thus, doing the decomposition of $\tilde{q}$ and mapping back, we get
where 
\begin{align*}
\hat{q}_0(k) &= \frac{L}{2}e^{-ik\frac{L}{2}}\int_{-1}^1 e^{-ixk\frac{L}{2}}\tilde{q}(x)\mathrm{d}x = \frac{L}{2}e^{-ik\frac{L}{2}}\left(\sum_{j=0}^{N-1} -\left(\frac{2}{ikL}\right)^{j+1}\left[\tilde{q}^{(j)}(x)e^{-ik\frac{L}{2}x}\right]_{-1}^1 + \left(\frac{2}{ikL}\right)^{N}\int_{-1}^1\tilde{q}^{(N)}(x)e^{-ik\frac{L}{2}x}\mathrm{d}x\right).\\
&= -\sum_{j=0}^{N-1} \left(\frac{2}{L}\right)^{j}\left(\frac{1}{ik}\right)^{j+1}\left[\tilde{q}^{(j)}(x)e^{-ik\frac{L}{2}(x+1)}\right]_{-1}^1 + \left(\frac{2}{L}\right)^{N-1}\left(\frac{1}{ik}\right)^{N}\int_{-1}^1\tilde{q}^{(N)}(x)e^{-ik\frac{L}{2}(x+1)}\mathrm{d}x
\end{align*}
Then define
$$
I_m(x,t) = \int_{-\infty}^{\infty}e^{ik x - \omega(k)t}(ik)^{-m}\mathrm{d}k
$$
We then have that the integral becomes
$$
I = \sum_{j=0}^{N-1} \left(\frac{2}{L}\right)^{j}\left(\tilde{q}^{(j)}(-1)I_{j+1}(x,t)-\tilde{q}^{(j)}(1)I_{j+1}(x-L,t)\right) + \left(\frac{2}{L}\right)^{N-1}\int_{-∞}^∞\frac{e^{ikx-\omega(k)t}}{(ik)^{N}}\int_{-1}^{1}\tilde{q}^{(N)}(s)e^{-ik\frac{L}{2}(s+1)}\mathrm{d}s\mathrm{d}k.
$$



# The latter integral
Let us now talk about this second integral
$$\int_{-∞}^∞\frac{e^{ikx-\omega(k)t}}{(ik)^{N}}\int_{-1}^{1}\tilde{q}^{(N)}(s)e^{-ik\frac{L}{2}(s+1)}\mathrm{d}s\mathrm{d}k$$
We aim to use numerically approximate this well. Switching the integrals and using the numerical approximation to get the integral $I_m$ here gave bad results for small $t ∼ 1e-5$. So now we look at Levins method. First things first, we split the integral $\int_{-∞}^∞\mathrm{d}k=\int_{-∞}^R+\int_{C}+\int_{R}^∞\mathrm{d}k$, where $C=\{Re^{i(π-t)}|t\in[0,π]\}$. Over $\int_{C}\mathrm{d}k$ we need to do God knows what. Let's hope Clenshaw-Curtis is enough. Over $\int_{R}^∞\mathrm{d}k$ and the other one, we hope to do Levin's method. For this, let's look at the Fourier transform first. Let's think about this in two ways. First, the naive way:
Say that 
$$ 
p'(x) - ik\frac{L}{2}p(x) = \tilde{q}^{(N)}(x)e^{-ik\frac{L}{2}},~~~x\in[-1,1].
$$
Solving this ODE standard ODE now gives us that
$$
\int_{-1}^{1}\tilde{q}^{(N)}(s)e^{-ik\frac{L}{2}(s+1)}\mathrm{d}s = p(1)e^{-ik\frac{L}{2}s} - p(-1)e^{ik\frac{L}{2}s}.
$$
This is one way to do the Fourier transform. Another is by making a bump function $ϕ(x)$ that filters out the high frequency (-1 is low, and 1 is high). We then have that we can write
\begin{align*}
\int_{-1}^{1}\tilde{q}^{(N)}(s)e^{-ik\frac{L}{2}(s+1)}\mathrm{d}s &= \int_{-1}^{1}\phi(s)e^{-ik\frac{L}{2}(s+1)}\tilde{q}^{(N)}(s)\mathrm{d}s+e^{-ik\frac{L}{2}}\int_{-1}^{1}(1-\phi(s))e^{-ik\frac{L}{2}s}\tilde{q}^{(N)}(s)\mathrm{d}s\\
&= \tilde{q}^{(N)}_+ + e^{-ikL}\int_{-1}^{1}(1-\phi(s))e^{-ik\frac{L}{2}(s-1)}\tilde{q}^{(N)}(s)\mathrm{d}s\\
& = \tilde{q}^{(N)}_+ + e^{-ikL}\tilde{q}^{(N)}_-
\end{align*}
Using this decomposition, we should be able to compute the first part of the integral:
$$
\int_{R}^∞\mathrm{d}k\frac{e^{ikx-\omega(k)t}}{(ik)^{N}}\left(\tilde{q}^{(N)}_+ + e^{-ikL}\tilde{q}^{(N)}_-\right)\mathrm{d}k=\int_{R}^∞\mathrm{d}k\frac{e^{ikx-\omega(k)t}}{(ik)^{N}}\tilde{q}^{(N)}_+\mathrm{d}k + \int_{R}^∞\mathrm{d}k\frac{e^{ik(x-L)-\omega(k)t}}{(ik)^{N}}\tilde{q}^{(N)}_-\mathrm{d}k.
$$
Here, one could approximate $\tilde{q}^{(N)}_+$ and $\tilde{q}^{(N)}_-$ with whatever means necesary. We now aim to solve two ODEs
$$
P_+'(k)+ixP_+(k)=\frac{e^{-\omega(k)t}}{(ik)^N}\tilde{q}^{(N)}_+ ~~~k∈ [R,\infty),
$$
and 
$$
P_-'(k)+i(x-L)P_-(k)=\frac{e^{-\omega(k)t}}{(ik)^N}\tilde{q}^{(N)}_- ~~~k∈ [R,\infty).
$$
For both we could put boundary condition $p_±(\infty)=0$ and truncate the domain to use Levin's method, or one could use mobius transformations. Say we have mobius Transformation mapping $[R,\infty) → [-1,1]$:
$ u =\frac{k-A}{k-B}$, with $2R-A=B$. We then have that $k=\frac{Bu+B-2R}{u-1}$. Long story short, we aim to solve for $Q_±(u) = P_±(k(u))$, in this case $Q'_± = P_\pm' k'$. We then get 
$$
\frac{(u-1)^2}{2(B-R)}Q_+'(u)+ixQ_+(u)=\frac{e^{-\omega(k(u))t}}{i^N}\left(\frac{u-1}{Bu+B-2R}\right)^N\tilde{q}^{(N)}_+ ~~~k∈ [R,\infty),
$$
and
$$
\frac{(u-1)^2}{2(B-R)}Q_-'(u)+i(x-L)Q_-(u)=\frac{e^{-\omega(k(u))t}}{i^N}\left(\frac{u-1}{Bu+B-2R}\right)^N\tilde{q}^{(N)}_- ~~~k∈ [R,\infty),
$$
where $Q_\pm(1) = 0$. We must also have that the root of $Bu+B-2R$ is not in $[-1,1]$. This means that, the smaller $R$ gets, the smaller $B$ must get. Let's chose $R=1$ and $B = .5$. Say we choose $N=2$, we can then simplify things to
$$
Q_+'(u)-\frac{ix}{(u-1)^2}Q_+(u)=\frac{4e^{-\omega(k(u))t}}{(u-3)^2}\tilde{q}^{(N)}_+ ~~~u∈ [-1,1],
$$
and
$$
Q_-'(u)-\frac{i(x-L)}{(u-1)^2}Q_-(u)=\frac{4e^{-\omega(k(u))t}}{(u-3)^2}\tilde{q}^{(N)}_- ~~~u∈ [-1,1].
$$
Here it comes in handy that we impose that $Q_\pm(1) = 0$. Note also that for both $\tilde{q}^{(N)}_+$ and $\tilde{q}^{(N)}_-$ are muffled! As a reminder, with $\phi(s)$ one for $s$ near -1 and zero near 1:
$$\tilde{q}^{(N)}_+ = \int_{-1}^{1}\phi(s)e^{-ik\frac{L}{2}(s+1)}\tilde{q}^{(N)}(s)\mathrm{d}s~~~\text{ and }~~~\tilde{q}^{(N)}_-=\int_{-1}^{1}(1-\phi(s))e^{-ik\frac{L}{2}(s-1)}\tilde{q}^{(N)}(s)\mathrm{d}s
$$
We hope enough dissipation will come from $\frac{4e^{-\omega(k(u))t}}{(u-3)^2}$ to counteract the increasing frequency with decay.

# Heat equation with left non-zero dirichlet boundary condition and zero initial condition.
Consider the heat equation with zero initial conidition and left non-zero dirichlet boundary condition, here we have $\omega(k)=k^2$ and $h_0≡0$:
\begin{align*}
q_t-\partial_{xx}q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= 0, &x\in[0,L],\\
q(0,t) &= g_0(t), & t ∈ [0,T],\\ 
q(L,t) &= 0, & t ∈ [0,T]
\end{align*}
Assume $\tilde{g}_0(k)=\int_0^Te^{ks}g_0(s)\mathrm{d}s$, we still need to find $\tilde{g}_1$ and $\tilde{h}_1$. We again built a system of equations through the global relation:
\begin{align*}
-ik\tilde{g}_0(k^2,t)-\tilde{g}_1(k^2,t)+e^{-ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(k,t)\\
ik\tilde{g}_0(k^2,t)-\tilde{g}_1(k^2,t)+e^{ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(-k,t)\\
\end{align*}
From this we get to solve the following system
$$
\begin{bmatrix}-1 & e^{-ikL}\\ -1 & e^{ikL}\end{bmatrix}\begin{bmatrix}\tilde{g}_1(k^2,t)\\ \tilde{h}_1(k^2,t)\end{bmatrix} = \begin{bmatrix}e^{k^2t}\hat{q}(k,t) + ik\tilde{g}_0(k^2,t)\\ e^{k^2t}\hat{q}(-k,t)-ik\tilde{g}_0(k^2,t)\end{bmatrix}.
$$
Solving this gives, remind yourself that $\Delta(k)=e^{-ikL}-e^{ikL}$:
$$
\begin{bmatrix}\tilde{g}_1(k^2,t)\\ \tilde{h}_1(k^2,t)\end{bmatrix} = \frac{1}{Δ(k)}\begin{bmatrix}e^{ikL} & -e^{-ikL}\\ 1 & -1\end{bmatrix} \begin{bmatrix}e^{k^2t}\hat{q}(k,t) + ik\tilde{g}_0(k^2,t)\\ e^{k^2t}\hat{q}(-k,t)-ik\tilde{g}_0(k^2,t)\end{bmatrix}.
$$
Therefore,
\begin{align*}
\tilde{g}_1(k^2,t)=ik\frac{e^{ikL}+e^{-ikL}}{\Delta(k)}\tilde{g}_0(k^2,t)+\frac{e^{ikL}+e^{-ikL}}{\Delta}\left(e^{k^2t}\hat{q}(-k,t)e^{-ikL}-e^{k^2t}\hat{q}(k)e^{ikL}\right),\\
\tilde{h}_1(k^2,t)=\frac{2ik}{\Delta(k)}\tilde{g}_0(k^2,t)+\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)-e^{k^2t}\hat{q}(k)\right).
\end{align*}
You can again convince yourself to ditch the fourier transform of the solution to get that,
\begin{align*}
&\tilde{g}(k,t)=ik\tilde{g}_0(k^2,t)+\tilde{g}_1(k^2,t)=ik\left(1+\frac{e^{ikL}+e^{-ikL}}{\Delta(k)}\right)\tilde{g}_0(k^2,t) = \frac{2ik}{1-e^{2ikL}}\tilde{g}_0(k^2,t),\\
&\tilde{h}(k,t)=\tilde{h}_1(k^2,t)=\frac{2ike^{-ikL}}{e^{-2ikL}-1}\tilde{g}_0(k^2,t).
\end{align*}
Therefore, the solution is given by
\begin{align*}
q(x,t)  &= \frac{1}{π}\int_{∂ D^+}e^{ikx-\omega(k)t}\frac{ik}{e^{2ikL}-1}\tilde{g}_0(k^2,t)\mathrm{d}k + \frac{1}{π}\int_{∂ D^-}e^{-ik(2L-x)-\omega(k)t}\frac{ik}{1-e^{-2ikL}}\tilde{g}_0(k^2,t)\mathrm{d}k.
\end{align*}

# The first boundary integral
We aim to compute
$$ I = \int_{∂ D^+}e^{ikx-\omega(k)t}\frac{ik}{e^{2ikL}-1}\int_0^t e^{\omega(k)s}g_0\mathrm{d}s\mathrm{d}k $$
Now, I'm gonna substitute in $\omega$, but it's easy to go down further with it. Note that is already nice to compute in the UHP, funny how that works. Note that if we have
$$
\overline{g}_0(s) = g_0(φ^{-1}(s)), \text{ with }\overline{g}_0:[-1,1]\to\mathbb{R}\text{ and } \varphi^{-1}:[-1,1]\to[0,t]: s ↦ \frac{t}{2}(s+1),
$$
then,
$$\tilde{g}_0(k,t) = \frac{t}{2}e^{\frac{t}{2}k}\tilde{\overline{g}}_0\left(\frac{t}{2}k,t\right).$$
$$
\tilde{g}_0(k^2,t) = \sum_{j=0}^{N-1}\left(-\frac{2}{t}\right)^j\left(\frac{1}{k}\right)^{2(j+1)}\left[e^{\frac{t}{2}k^2(s+1)}\overline{g}_0^{(j)}(s)\right]_{-1}^{1}-\left(-\frac{2}{t}\right)^{N-1}\left(\frac{1}{k}\right)^{2N}∫_{-1}^1e^{\frac{t}{2}k^2(s+1)}\overline{g}_0^{(N)}(s)\mathrm{d}s
$$
Now define
$$I_m(x,t) = \int_{∂ D^+}e^{ixk-\omega(k)t}\frac{i}{(e^{2ikL}-1)k^{m-1}}$$

# The second boundary integral
We aim to compute
$$ I = \int_{∂ D^+}e^{-ik(2L-x)-\omega(k)t}\frac{ik}{1-e^{-2ikL}}\int_0^t e^{\omega(k)s}g_0\mathrm{d}s\mathrm{d}k $$
What we really want to compute is:
$$ \int_{∂ D^+}e^{-ik(2L-x)-\omega(k)t}\frac{ik}{1-e^{-2ikL}}\int_0^t e^{\omega(k)s}g_0\mathrm{d}s\mathrm{d}k, $$
for $t>T$, since 
$$ \int_{∂ D^+}e^{\omega(k)(s-t)}\frac{ik}{1-e^{-2ikL}}e^{-ik(2L-x)}g_0(s)\mathrm{d}k=0~~~ ∀ s>t. $$
This as $s-t>0$ and Jordan's lemma. Therefore, 
$$  \int_{∂ D^+}e^{-ik(2L-x)-\omega(k)t}\frac{ik}{1-e^{-2ikL}}\int_t^T e^{\omega(k)s}g_0\mathrm{d}s\mathrm{d}k, $$
and thus,
$$ I = \int_{∂ D^+}e^{-ik(2L-x)-\omega(k)t}\frac{ik}{1-e^{-2ikL}}\int_0^T e^{\omega(k)s}g_0\mathrm{d}s\mathrm{d}k. $$

# Heat equation with right non-zero dirichlet boundary condition and zero initial condition.

$$
\begin{cases}
\partial_t u + \omega(i\partial_x)u=0,\\
u(x,0) = 0,\\
u(0,t) = 0,~u(L,t) = g(t),
\end{cases}
$$
through change of variables $\tilde{x}=L-x$ and then solve 
$$
\begin{cases}
\partial_t u + \tilde{\omega}(i\partial_{x})u=0,\\
u(x,0) = 0,\\
u(0,t) = g(t),~u(L,t) = 0,
\end{cases}
$$
where $\tilde{\omega} = \sum_{j=0}^n (-1)^nw_n k^{n}$.
