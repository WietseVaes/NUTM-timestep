# General UTM solution on the finite interval

Say we want to solve 
\begin{align*}
q_t+\omega(-i\partial_x)q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= q_0, &x\in[0,L],\\
\partial_x^j q(0,t) &= g_j(t), & j=0,\dots,n-1,\\
\partial_x^j q(L,t) &= h_j(t), & j=0,\dots,n-1,
\end{align*}
where $\omega(k)\in\mathbb{P}^{n}$. Note that the boundary conditions are not neccesarily imposed, but some should follow from others through a procedure that will be given by example.
*** Work out that the solution is given by ***

\begin{align*}
q(x,t)  = \frac{1}{2π}\int_{-\infty}^{\infty}e^{ikx-\omega(k)t}\hat{q}_0(k)\mathrm{d}k &- \frac{1}{2π}\int_{∂ D^+}e^{ikx-\omega(k)}\tilde{g}(k,t)\mathrm{d}k\\
& - \frac{1}{2π}\int_{∂ D^-}e^{-ik(L-x)-\omega(k)}\tilde{h}(k,t)\mathrm{d}k,
\end{align*}
where $\hat{q}_0=\int_{0}^{L}e^{-ikx}q_0\mathrm{d}x$, $∂ D^+\{k|\mathcal{I}m(k) ≥ 0 \& \omega(k)=0\}$ and $∂ D^-\{k|\mathcal{I}m(k) ≤ 0 \& \omega(k)=0\}$. We now also have $\tilde{g} = \sum_{j=0}^{n-1}c_j(k)\tilde{g}_j(\omega(k),t)$, with $\tilde{g}_j=\int_{0}^t e^{ks}g(s)\mathrm{d}s$ and respectively the same for $\tilde{h}$, we also define the coeficients through
$$ \sum_{j=0}^{n-1}c_j(k)\partial_x^jq(x,t) = i\left(\frac{\omega(k)-\omega(l)}{k-l}\right)\Big|_{l=-i\partial_x}q(x,t).$$

# Heat equation with homogeneous boundary condition.
To get started, let's deal with the heat equation with Homogeneous boundary conditions, here we have $\omega(k)=k^2$, $g_0≡0$ and $h_0≡0$:
\begin{align*}
q_t-\partial_{xx}q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= q_0, &x\in[0,L],\\
q(0,t) &= 0, & t ∈ [0,T],\\ 
q(L,t) &= 0, & t ∈ [0,T]
\end{align*}
We first need to find $\tilde{g}$ and $\tilde{h}$, but for this we need to find $g_1$ and $h_1$ and the coefficients $c_j(k)$. The latter is easy, since
$$c_0(k)q(x,t)+c_1(k)\partial_x q(x,t)=i\left(\frac{k^2-l^2}{k-l}\right)\Big|_{l=-i\partial_x}q(x,t)=ikq(x,t)+\partial_xq(x,t)~⇒~c_0 = ik~\&~c_1= 1.$$
The more difficult part is to find $g_1$ and $h_1$. For this purpose we use the global relation *** it's easy to get, but I'm gonna ignore that for now***
$$\hat{q}_0(k)-\tilde{g}(\omega(k),t)+e^{-ikL}\tilde{h}(\omega(k),t)=e^{\omega(k)t}\hat{q}(k,t),~~k\in\mathbb{C}.$$
Using invariances under $\omega$, the previous coefficients and $h_0≡ g_0≡ 0$, we get
\begin{align*}
\hat{q}_0(k)-\tilde{g}_1(k^2,t)+e^{ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(k,t)\\
\hat{q}_0(-k)-\tilde{g}_1(k^2,t)+e^{-ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(-k,t)\\
\end{align*}
Solving the system to get $\tilde{g}_1$ and $\tilde{h}_1$ gives us for $\Delta(k) = e^{ikL}-e^{-ikL}$,
\begin{align*}
\tilde{g}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right)-\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)e^{-ikL}-e^{k^2t}\hat{q}(k)e^{ikL}\right),\\
\tilde{h}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right)-\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)-e^{k^2t}\hat{q}(k)\right).
\end{align*}
*** Show that the latter of the two will eventually be equal to zero when we use the global relation and use the inverse fourier transform on it. This is true because of analyticity and the Jordan's lemma to prove that the whole upper (or lower) half plane is nicely behaving such that that arc is zero as it's radius goes to infinity.***
Therefore,
\begin{align*}
&\tilde{g}(k,t)=\tilde{g}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right),\\
&\tilde{h}(k,t)=\tilde{h}_1(k^2,t)=\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right).
\end{align*}
The solution is then of the form 
\begin{align*}
q(x,t)  = \frac{1}{2π}\int_{-\infty}^{\infty}e^{ikx-\omega(k)t}\hat{q}_0(k)\mathrm{d}k &- \frac{1}{2π}\int_{∂ D^+}e^{ikx-\omega(k)t}\frac{1}{\Delta}\left(\hat{q}_0(-k)e^{-ikL}-\hat{q}_0(k)e^{ikL}\right)\mathrm{d}k\\
& - \frac{1}{2π}\int_{∂ D^-}e^{-ik(L-x)-\omega(k)t}\frac{1}{\Delta}\left(\hat{q}_0(-k)-\hat{q}_0(k)\right)\mathrm{d}k.
\end{align*}
Numerical approximations go as follows, note that we put an emphasis on the asymptotics for $t→0$ as we aim to build a timestepping method.

# Heat equation with left non-zero dirichlet boundary condition and zero initial condition.
Consider the heat equation with zero initial conidition and left non-zero dirichlet boundary condition, here we have $\omega(k)=k^2$ and $h_0≡0$:
\begin{align*}
q_t-\partial_{xx}q&=0,&(x,t)\in[0,L]×[0,T],\\
q(x,0) &= 0, &x\in[0,L],\\
q(0,t) &= g_0(t), & t ∈ [0,T],\\ 
q(L,t) &= 0, & t ∈ [0,T]
\end{align*}
Assume $\tilde{g}_0(k)=\int_0^Te^{ks}g_0(s)\mathrm{d}s$, we still need to find $\tilde{g}_1$ and $\tilde{h}_1$. We again built a system of equations through the global relation:
\begin{align*}
-ik\tilde{g}_0(k^2,t)-\tilde{g}_1(k^2,t)+e^{ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(k,t)\\
ik\tilde{g}_0(k^2,t)-\tilde{g}_1(k^2,t)+e^{-ikL}\tilde{h}_1(k^2,t)&=e^{k^2t}\hat{q}(-k,t)\\
\end{align*}
From this we get to solve the following system
$$
\begin{bmatrix}-1 & e^{ikL}\\ -1 & e^{-ikL}\end{bmatrix}\begin{bmatrix}\tilde{g}_1(k^2,t)\\ \tilde{h}_1(k^2,t)\end{bmatrix} = \begin{bmatrix}e^{k^2t}\hat{q}(k,t) + ik\tilde{g}_0(k^2,t)\\ e^{k^2t}\hat{q}(k,t)-ik\tilde{g}_0(k^2,t)\end{bmatrix}.
$$
Solving this gives, remind yourself that $\Delta(k)=e^{ikL}-e^{-ikL}$:
$$
\begin{bmatrix}\tilde{g}_1(k^2,t)\\ \tilde{h}_1(k^2,t)\end{bmatrix} = \frac{1}{Δ(k)}\begin{bmatrix}e^{ikL} & -e^{-ikL}\\ 1 & -1\end{bmatrix} \begin{bmatrix}e^{k^2t}\hat{q}(k,t) + ik\tilde{g}_0(k^2,t)\\ e^{k^2t}\hat{q}(k,t)-ik\tilde{g}_0(k^2,t)\end{bmatrix}.
$$
Therefore,
\begin{align*}
\tilde{g}_1(k^2,t)=ik\frac{e^{ikL}+e^{-ikL}}{\Delta(k)}\tilde{g}_0(k^2,t)-\frac{e^{ikL}+e^{-ikL}}{\Delta}\left(e^{k^2t}\hat{q}(-k,t)e^{-ikL}-e^{k^2t}\hat{q}(k)e^{ikL}\right),\\
\tilde{h}_1(k^2,t)=\frac{2ik}{\Delta(k)}\tilde{g}_0(k^2,t)-\frac{1}{\Delta(k)}\left(e^{k^2t}\hat{q}(-k,t)-e^{k^2t}\hat{q}(k)\right).
\end{align*}
You can again convince yourself to ditch the fourier transform of the solution to get that,
\begin{align*}
&\tilde{g}(k,t)=ik\tilde{g}_0(k^2,t)+\tilde{g}_1(k^2,t)=ik\left(1+\frac{e^{ikL}+e^{-ikL}}{\Delta(k)}\right)\tilde{g}_0(k^2,t) = \frac{ike^{2ikL}}{e^{2ikL}-1}\tilde{g}_0(k^2,t),\\
&\tilde{h}(k,t)=\tilde{h}_1(k^2,t)=\frac{2ike^{-ikL}}{1-e^{-2ikL}}\tilde{g}_0(k^2,t).
\end{align*}
Therefore, the solution is given by
\begin{align*}
q(x,t)  &= - \frac{1}{2π}\int_{∂ D^+}e^{ik(x+2L)-\omega(k)t}\frac{ik}{e^{2ikL}-1}\tilde{g}_0(k^2,t)\mathrm{d}k - \frac{1}{2π}\int_{∂ D^-}e^{-ik(2L-x)-\omega(k)t}\frac{2ik}{1-e^{-2ikL}}\tilde{g}_0(k^2,t)\mathrm{d}k.
\end{align*}
