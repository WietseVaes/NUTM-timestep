```julia
w = [0,0,0,-1]
N = 30;
x = 0
t = .00001
funt = N -> SpecialFunction(w, x, t, -1, N)
funt(N)|>display
refs = funt(400)
plot(2:400,abs.(real.(funt.(collect(2:400)) .- refs)) .+ eps(), xlabel = "N", ylabel = "integral erre", label = "Real", yaxis =:log)
plot!(2:400,abs.(imag.(funt.(collect(2:400)) .- refs)) .+ eps(), label = "Imaginairy")
```
